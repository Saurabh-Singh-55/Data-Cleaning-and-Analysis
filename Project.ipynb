```r
num_rows <- sum(!grepl("cm", data$Height, fixed = TRUE))
cat(paste("There are", num_rows, "rows not containing \"cm\"."))
height_counts <- table(data$Height[!grepl("cm", data$Height, fixed = TRUE)])
print(height_counts)
```

```r
library(stringr)
convertHeight <- function(height) {
  if (!grepl("'", height, fixed = TRUE)) {
    return(height)
  }
  height_parts <- str_extract_all(height, "\\d+")

  feet <- as.numeric(height_parts[[1]][1])
  inch <- as.numeric(height_parts[[1]][2])
  return(as.character(round(feet * 30.48 + inch * 2.54)))
}
print("check for converstion 5feet 9inch")
convertHeight("5'10\"")
# Apply the convertHeight function to the data$Height column
data$Height <- sapply(data$Height, convertHeight)

# Remove "cm" and convert height to numeric
data$Height <- gsub("cm", "", data$Height)
data$Height <- as.numeric(data$Height)

# Print the resulting data$Height column
print(table(data$Height))
```

```r
library(stringr)

convertWeight <- function(weight) {
  if (!grepl("lbs", weight, fixed = TRUE)) {
    return(weight)
  }
  weight_val <- as.numeric(str_extract(weight, "\\d+"))
  return(as.character(round(weight_val * 0.45359237, 0)))
}

print("check for converstion lbs to kg 130 lbs")
convertWeight("130lbs")
# Apply the convertWeight function to the data$Weight column
data$Weight <- sapply(data$Weight, convertWeight)

# Remove "lbs" and convert weight to numeric
data$Weight <- gsub("kg", "", data$Weight)

data$Weight <- as.numeric(data$Weight)

# Print the resulting data$Weight column
print(table(data$Weight))
```

```r
#convert money format k in thousand and m in million value.

convertMoney <- function(money) {
  value <- as.numeric(gsub("[^0-9.]", "", money))
  if (is.na(value)) value <- 0
  if (grepl("M", money)) {
    value <- value * 1000000
  } else if (grepl("K", money)) {
    value <- value * 1000
  }
  return(as.character(as.integer(value)))
}

data$Value <- as.numeric(sapply(data$Value, convertMoney))
data$Wage <- as.numeric(sapply(data$Wage, convertMoney))
```

## Sampling

Draw various random samples (using at least 3 different sample sizes) of the
data and show the applicability of the Central Limit Theorem for at least one
variable. (15 points)

```r
install.packages("sampling")
library(sampling)
```

```r
#Draw various random samples (using at least 3 different sample sizes)
# of the data and show the applicability of the Central Limit Theorem for at least one

sample_mean <- function(x, nsamples, sample_size){
    Meanx <- numeric(nsamples)
    for (i in 1:nsamples) {
        Meanx[i] <- mean(sample(x,sample_size, replace=TRUE))
    }
    return(Meanx)
}


xdata <- data$Height
#sampling 10000,  sample size = 5
Meanx <- sample_mean(xdata, 5000, 5)
hist(Meanx, prob = TRUE, breaks = 15, main = "if Sample Size is 5")
curve(dnorm(x, mean=mean(Meanx),sd=sd(Meanx)), add=TRUE, col="red", lwd=2)



#sampling 10000,  sample size = 100
Meanx <- sample_mean(xdata, 5000, 1000)
hist(Meanx, prob = TRUE, breaks = 15, main = "if Sample Size is 1000")
curve(dnorm(x, mean=mean(Meanx),sd=sd(Meanx)), add=TRUE, col="red", lwd=2)



#sampling 10000, sample size = 10000
Meanx <- sample_mean(xdata, 5000, 5000)
hist(Meanx, prob = TRUE, breaks = 15, main = "if Sample Size is 10000")
curve(dnorm(x, mean=mean(Meanx),sd=sd(Meanx)), add=TRUE, col="red", lwd=2)
```

Simple random without replace

```r
# Simple Random Sampling with replacement
sample_size <- 100  # Replace with desired sample size
simple_random_sample <- data[sample(1:nrow(data), size = sample_size, replace = FALSE), ]
# head(simple_random_sample)
dim(data)
dim(simple_random_sample)
```

Conclusions: By using simple random sampling, we can obtain a representative sample of the entire dat population. The conclusions drawn from this sample can be generalized to the whole population with a certain level of confidence. However, it's important to consider the potential variability introduced by the randomness of the sample.

Simple random with Replacement

```r
sample_size <- 100  # Replace with desired sample size
simple_random_sample <- data[sample(1:nrow(data), size = sample_size, replace = TRUE), ]
# head(simple_random_sample)
dim(data)
dim(simple_random_sample)
```

Conclusions

•The sample may contain duplicate elements from the dataset, potentially affecting the representation of certain elements in the sample.

•Statistical inferences made from the sample need to take into account the potential presence of duplicates and the impact on variability estimation.

•The sample size may be the same as or larger than the whole dataset, as elements can be selected multiple times.

•If the sample is significantly larger than needed, it may introduce unnecessary computational overhead without providing additional benefits.

Stratified sampling

```r
# Stratified Sampling
stratum_sample_size <- 50  # Replace with desired sample size for each stratum

strata <- unique(data$Nationality)
stratified_sample <- data.frame()

for (s in strata) {
  stratum_data <- data[data$Nationality == s, ]
  stratum_sample <- stratum_data[sample(1:nrow(stratum_data), size = stratum_sample_size, replace = FALSE), ]
  stratified_sample <- rbind(stratified_sample, stratum_sample)
}
table(data$Nationality)
table(stratified_sample$Nationality)
```

**•Improved Representation:** Stratified sampling is used to ensure that each subgroup or stratum within the population is adequately represented in the sample. By dividing the population into homogeneous subgroups based on relevant characteristics, stratified sampling captures the diversity and variability of the population. This leads to more accurate and reliable estimates and inferences for each subgroup and the overall population.

**•Increased Precision:** Stratified sampling allows for more precise estimates and inferences compared to using the whole dataset or simple random sampling. By allocating sample size proportionally to each stratum based on its variability, stratified sampling increases the precision of estimates within each subgroup. This is particularly valuable when there are significant variations or heterogeneity within the population, as it enables more focused analysis and precise conclusions.

**•Targeted Analysis:** Stratified sampling enables the analysis of specific subgroups within the population. By dividing the population into strata, researchers can focus their analysis on each subgroup individually, allowing for targeted examination of their characteristics, trends, or patterns. This targeted analysis provides deeper insights into the specific subgroups of interest, facilitating more nuanced conclusions and informed decision-making.
